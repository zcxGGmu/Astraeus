&emsp;&emsp;è¿™é‡Œéœ€è¦å…³æ³¨çš„å…³é”®è¿æ¥ä¿¡æ¯æ˜¯ï¼š`Redis`çš„é»˜è®¤è¿æ¥åœ°å€æ˜¯`127.0.0.1`ï¼Œç«¯å£å·æ˜¯`6379`ï¼Œæ•°æ®åº“æ˜¯`0`ã€‚åœ¨ `Redis` ä¸­ï¼Œæ•°æ®åº“ 0 æ˜¯é»˜è®¤çš„æ•°æ®åº“ã€‚`Redis` é»˜è®¤æä¾›äº† 16 ä¸ªé€»è¾‘æ•°æ®åº“ï¼Œç¼–å·ä» 0 åˆ° 15ã€‚æ¯ä¸ªæ•°æ®åº“éƒ½æ˜¯ç‹¬ç«‹çš„ï¼Œå…·æœ‰è‡ªå·±çš„é”®ç©ºé—´ã€‚

````python
# pip install redis  # å®‰è£…redisåº“

import redis

r = redis.Redis(host='localhost', port=6379, db=0)
````

&emsp;&emsp;å»ºç«‹è¿æ¥åï¼Œæˆ‘ä»¬å°±å¯ä»¥å¯¹`Redis`è¿›è¡Œæ“ä½œäº†ã€‚ ç›¸å…³çš„å¢åˆ æ”¹æŸ¥ç¤ºä¾‹éƒ½éå¸¸ç®€å•ï¼Œä»£ç å¦‚ä¸‹ï¼š


````python
# å¢ï¼šæ·»åŠ æ•°æ®
def add_data(key, value):
    r.set(key, value)           # setï¼šè®¾ç½®é”®å€¼å¯¹
    print(f"æ·»åŠ æ•°æ®: {key} -> {value}")

# æŸ¥ï¼šè·å–æ•°æ®
def get_data(key):
    value = r.get(key)         # getï¼šè·å–é”®å€¼å¯¹
    if value is not None:
        print(f"è·å–æ•°æ®: {key} -> {value.decode('utf-8')}")
    else:
        print(f"é”® {key} ä¸å­˜åœ¨")

# æ”¹ï¼šæ›´æ–°æ•°æ®
def update_data(key, new_value):
    if r.exists(key):           # existsï¼šåˆ¤æ–­é”®æ˜¯å¦å­˜åœ¨
        r.set(key, new_value)   # setï¼šè®¾ç½®é”®å€¼å¯¹ï¼Œå¦‚æœé”®å­˜åœ¨ï¼Œåˆ™æ›´æ–°é”®å€¼å¯¹
        print(f"æ›´æ–°æ•°æ®: {key} -> {new_value}")
    else:
        print(f"é”® {key} ä¸å­˜åœ¨ï¼Œæ— æ³•æ›´æ–°")

# åˆ ï¼šåˆ é™¤æ•°æ®
def delete_data(key):
    if r.delete(key):    # deleteï¼šåˆ é™¤é”®å€¼å¯¹
        print(f"åˆ é™¤æ•°æ®: {key}")
    else:
        print(f"é”® {key} ä¸å­˜åœ¨ï¼Œæ— æ³•åˆ é™¤")
````

&emsp;&emsp;å¯ä»¥è°ƒç”¨`add_data`ã€`get_data`ã€`update_data`ã€`delete_data`å‡½æ•°æ¥å¯¹`Redis`è¿›è¡Œæ“ä½œã€‚

````python
# æ·»åŠ æ•°æ®
add_data("name", "muyu")
add_data("age", "30")
````

&emsp;&emsp;è¿™é‡Œéœ€è¦å…³æ³¨çš„å…³é”®è¿æ¥ä¿¡æ¯æ˜¯ï¼š`Redis`çš„é»˜è®¤è¿æ¥åœ°å€æ˜¯`127.0.0.1`ï¼Œç«¯å£å·æ˜¯`6379`ï¼Œæ•°æ®åº“æ˜¯`0`ã€‚åœ¨ `Redis` ä¸­ï¼Œæ•°æ®åº“ 0 æ˜¯é»˜è®¤çš„æ•°æ®åº“ã€‚`Redis` é»˜è®¤æä¾›äº† 16 ä¸ªé€»è¾‘æ•°æ®åº“ï¼Œç¼–å·ä» 0 åˆ° 15ã€‚æ¯ä¸ªæ•°æ®åº“éƒ½æ˜¯ç‹¬ç«‹çš„ï¼Œå…·æœ‰è‡ªå·±çš„é”®ç©ºé—´ã€‚

````python
# pip install redis  # å®‰è£…redisåº“

import redis

r = redis.Redis(host='localhost', port=6379, db=0)
````

&emsp;&emsp;å»ºç«‹è¿æ¥åï¼Œæˆ‘ä»¬å°±å¯ä»¥å¯¹`Redis`è¿›è¡Œæ“ä½œäº†ã€‚ ç›¸å…³çš„å¢åˆ æ”¹æŸ¥ç¤ºä¾‹éƒ½éå¸¸ç®€å•ï¼Œä»£ç å¦‚ä¸‹ï¼š


````python
# å¢ï¼šæ·»åŠ æ•°æ®
def add_data(key, value):
    r.set(key, value)           # setï¼šè®¾ç½®é”®å€¼å¯¹
    print(f"æ·»åŠ æ•°æ®: {key} -> {value}")

# æŸ¥ï¼šè·å–æ•°æ®
def get_data(key):
    value = r.get(key)         # getï¼šè·å–é”®å€¼å¯¹
    if value is not None:
        print(f"è·å–æ•°æ®: {key} -> {value.decode('utf-8')}")
    else:
        print(f"é”® {key} ä¸å­˜åœ¨")

# æ”¹ï¼šæ›´æ–°æ•°æ®
def update_data(key, new_value):
    if r.exists(key):           # existsï¼šåˆ¤æ–­é”®æ˜¯å¦å­˜åœ¨
        r.set(key, new_value)   # setï¼šè®¾ç½®é”®å€¼å¯¹ï¼Œå¦‚æœé”®å­˜åœ¨ï¼Œåˆ™æ›´æ–°é”®å€¼å¯¹
        print(f"æ›´æ–°æ•°æ®: {key} -> {new_value}")
    else:
        print(f"é”® {key} ä¸å­˜åœ¨ï¼Œæ— æ³•æ›´æ–°")

# åˆ ï¼šåˆ é™¤æ•°æ®
def delete_data(key):
    if r.delete(key):    # deleteï¼šåˆ é™¤é”®å€¼å¯¹
        print(f"åˆ é™¤æ•°æ®: {key}")
    else:
        print(f"é”® {key} ä¸å­˜åœ¨ï¼Œæ— æ³•åˆ é™¤")
````

&emsp;&emsp;å¯ä»¥è°ƒç”¨`add_data`ã€`get_data`ã€`update_data`ã€`delete_data`å‡½æ•°æ¥å¯¹`Redis`è¿›è¡Œæ“ä½œã€‚

````python
# æ·»åŠ æ•°æ®
add_data("name", "muyu")
add_data("age", "30")
````

````python
# æŸ¥æ•°æ®
get_data("name")
get_data("age")
````

````python
# æ”¹æ•°æ®
update_data("age", "31")

# æŸ¥æ›´æ–°åçš„æ•°æ®
get_data("age")
````

````python
# åˆ æ•°æ®
delete_data("name")

# æŸ¥å·²åˆ é™¤çš„æ•°æ®
get_data("name")
````
